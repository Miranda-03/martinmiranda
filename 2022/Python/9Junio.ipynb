```python

import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import matplotlib.pyplot as plt3
import matplotlib.pyplot as plt4

df1 = pd.read_csv("sample_data/registro-nacional-sociedades-muestreo.csv")
df1 = df1.loc[0:500]


x_7 = df1['dom_fiscal_estado_domicilio'].unique()
#x_7 = x_7[x_7 != nan]

y_8 = df1['dom_fiscal_estado_domicilio'].value_counts().tolist()
y_8.reverse()

df2 = df1

x_1 = df1['tipo_societario'].unique()
y_2 = df1['tipo_societario'].value_counts().tolist()
y_2.reverse()

df2 = df2.loc[0:10]
x_3 = df2['dom_legal_localidad'].unique()
y_4 = df2['dom_legal_localidad'].value_counts().tolist()
y_4.reverse()

df3 = df1
x_5 = df2['dom_legal_provincia'].unique()
y_6 = df2['dom_legal_provincia'].value_counts().tolist()
y_6.reverse()

# grfico 1
plt1.figure(figsize=(30, 10))
plt1.bar(x_1,y_2 )
plt1.title('Sociedades creadas')
plt1.xlabel('Sociedades')
plt1.ylabel('cantidad de sociedades')
plt1.show()

# grfico 2
plt2.figure(figsize=(20, 10))
plt2.bar(x_3,y_4)
plt2.title('Sociedades creadas en provincias')
plt2.xlabel('Provincias')
plt2.ylabel('cantidad de Sociedades')
plt2.show()

# grfico 2
plt3.figure(figsize=(20, 10))
plt3.bar(x_5,y_6)
plt3.title('Sociedades creadas en localidades')
plt3.xlabel('Localidades')
plt3.ylabel('cantidad de sociedades')
plt3.show()

# grfico 2
plt4.figure(figsize=(10, 10))
plt4.bar(x_7,y_8)
plt4.title('Sociedades declaradas')
plt4.xlabel('tipos de declaraciones')
plt4.ylabel('cantidad de sociedades')
plt4.show()
```
