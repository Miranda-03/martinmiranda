```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

df1 = pd.read_csv("sample_data/dataset_viajes_sube.csv")
df1 = df1.loc[0:300]
x_values = df1['TIPO_TRANSPORTE'].unique()
y_values = df1['TIPO_TRANSPORTE'].value_counts().tolist()
y_values.reverse()

# gráfico 1
plt1.bar(x_values,y_values)
plt1.title('Viajes segun vehículos')
plt1.xlabel('viajes con tarjeta sube')
plt1.ylabel('cantidad de viajes')
plt1.show()

# gráfico 2
plt2.plot(x_values,y_values);
plt2.title('Viajes segun vehículos')
plt2.xlabel('viajes con tarjeta sube')
plt2.ylabel('cantidad de viajes')
plt2.show()
```

1) Mi conclusión es que la tarjeta sube se utiliza mas en los colectivos.

2) Me parecería importante saber la cantidad de personas que viajan y no los viajes que se hacen con la sube para tener mas precisión.

3) Creo que combinando las dos columnas con un promedio de las personas que viajan por dia.
